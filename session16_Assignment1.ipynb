# Casting the Variables according to the respective data type

```python
def cast(data_type, value):
    if data_type == 'INT':
        return int(value)
    elif data_type == 'DATE':
        return datetime.strptime(value, '%m/%d/%Y')
    else:
        return str(value)

def cast_row(data_types, data_row):
    return [cast(data_type, value) 
            for data_type, value in zip(data_types, data_row)]
```

Iterator for employment

```python
def read_emp_file(file_name):
    with open(file_name) as f:
        rows = csv.reader(f, delimiter=',', quotechar='"')
        yield from rows
```

```python
file1_iter = read_emp_file('employment.csv')
headers = next(file1_iter)
print(headers)
employment = namedtuple('employment',headers)
data_types = ['STRING','STRING','STRING','STRING']
employees = (employment(*cast_row(data_types, 
                          line))
            for line in file1_iter)
```

Goal 4
=========
Find the largest group of car makes for each gender.

Possibly more than one such group per gender exists (equal sizes).

```python
male_dict={}
female_dict={}
for data in consolidated_info:
    if data.gender == 'Male':
        if data.vehicle_make in male_dict.keys():
            male_dict[data.vehicle_make] +=1
        else:
            male_dict[data.vehicle_make] =1
    else:
        if data.vehicle_make in female_dict.keys():
            female_dict[data.vehicle_make] +=1
        else:
            female_dict[data.vehicle_make] =1
        
            
```

```python
for w in sorted(male_dict, key=male_dict.get, reverse=True):
    print(w, male_dict[w])
```

```python
for w in sorted(female_dict, key=female_dict.get, reverse=True):
    print(w, female_dict[w])
```

```python
male_largest_car = max(male_dict, key= lambda x: male_dict[x]) 
female_largest_car = max(female_dict, key= lambda x: female_dict[x]) 
```
